```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load
schools_data_file  = "Resources/schools_complete.csv"
students_data_file = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
schools_data = pd.read_csv(schools_data_file)


students_data = pd.read_csv(students_data_file)

# Combine the data into a single dataset
complete_schools_data = pd.merge(students_data, schools_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
# Calculate the total number of schools
TotalSchools = complete_schools_data["school_name"].nunique()
# Calculate the total number of students
TotalStudents = complete_schools_data["Student ID"].nunique()
# Calculate the total budget
TotalBudget = schools_data["budget"].sum()
# Calculate the average math score
MathScoreaverage = complete_schools_data["math_score"].mean()
# Calculate the average reading score
ReadingScoreaverage = complete_schools_data["reading_score"].mean()
# Calculate the percentage of passing math score (70 or greater)
passingMath = len(complete_schools_data[complete_schools_data["math_score"] >= 70])
PassingMathpercentage = passingMath / TotalStudents * 100
# Calculate the percentage of passing reading score
passingReading = len(complete_schools_data[complete_schools_data["reading_score"] >= 70])
PassingReadingpercentage = passingReading / TotalStudents * 100
# Calculate the overall passing 
OverallPassingPercentage = (MathScoreaverage + ReadingScoreaverage) / 2

# Create a dataframe to hold the above results
district_summary_df = pd.DataFrame({
    "Total Schools": [TotalSchools],
    "Total Students": [TotalStudents],
    "Total Budget": [TotalBudget],
    "Average Math Score": [MathScoreaverage],
    "Average Reading Score": [ReadingScoreaverage],
    "% Passing Math": [PassingMathpercentage],
    "% Passing Reading": [PassingReadingpercentage],
    "% Overall Passing": [OverallPassingPercentage]
})

district_summary_df["Total Budget"] = district_summary_df["Total Budget"].map("${:,.2f}".format)

district_summary_df[["Total Schools",
      "Total Students",
      "Total Budget",
      "Average Math Score",
      "Average Reading Score",
      "% Passing Math",
      "% Passing Reading",
      "% Overall Passing"]]
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
# Grouped by school name 
grouped_school = complete_schools_data.groupby(['school_name'])
# Calculate total student
total_student_by_school = grouped_school.size()
total_student_by_school
# Collect school type
school_type = grouped_school['type'].first()
school_type
# Calculate total budget
total_budget = grouped_school['budget'].first()
total_budget
# Calculate budget per each student
t_budget_per_student = total_budget/total_student_by_school
t_budget_per_student
# Calculate the average math score
avg_math_by_school=grouped_school["math_score"].mean()
avg_math_by_school
# Calculate the average reading score
avg_reading_by_school=grouped_school["reading_score"].mean()
avg_reading_by_school
# Calculate passing math score percentange
grouped_passing_math = complete_schools_data[complete_schools_data['math_score']>=70].groupby(['school_name']).size()
percent_passing_math = (grouped_passing_math/TotalStudents)*100
percent_passing_math
# Calculate reading score percentange
grouped_passing_reading = complete_schools_data[complete_schools_data['reading_score']>=70].groupby(['school_name']).size()
percent_passing_reading = (grouped_passing_reading/TotalStudents)*100
percent_passing_reading
# Calculate the overall passing percentage
percent_overall_passing = (percent_passing_math + percent_passing_reading)/2
percent_overall_passing

# hold the above results 
school={
    'School Type': school_type,
    'Total Students':total_student_by_school,
    'Total School Budget':total_budget,
    'Per Student Budget':t_budget_per_student,
    'Average Math Score': avg_math_by_school,
    'Average Reading Score': avg_reading_by_school,
    '% Passing Math': percent_passing_math,
    '% Passing Reading': percent_passing_reading,
    '% Overall Passing': percent_overall_passing
}
school_summary = pd.DataFrame(school)

school_summary["Total School Budget"] = school_summary["Total School Budget"].map("${:,.2f}".format)
school_summary["Per Student Budget"] = school_summary["Per Student Budget"].map("${:,.2f}".format)

school_summary
```

## Top Performing Schools (By % Overall Passing)

* Sort and display the top five performing schools by % overall passing.

```python
# Create an table for schools
# School Name
school_symmary_df = complete_schools_data.groupby(["school_name"])
# School Type
schoolType = school_symmary_df["type"].first()
# Total Students
totalStudents = school_symmary_df["Student ID"].count()
# Total School Budget
totalSchoolBudget = school_symmary_df["budget"].first()
# Per Student Budget
perStudentBudget = totalSchoolBudget / totalStudents
# Average Math Score
MathScoreaverage = school_symmary_df["math_score"].mean()
# Average Reading Score
ReadingScoreaverage = school_symmary_df["reading_score"].mean()
# % Passing Math
passingMath = complete_schools_data[complete_schools_data["math_score"] >= 70].groupby(["school_name"])["math_score"].count()
PassingMathpercentage = passingMath / totalStudents * 100
# % Passing Reading
passingRead = complete_schools_data[complete_schools_data["reading_score"] >= 70].groupby(["school_name"])["reading_score"].count()
PassingReadingpercentage = passingRead / totalStudents * 100
# % Overall Passing
OverallPassingPercentage = (PassingMathpercentage + PassingReadingpercentage) / 2

# hold the above results
school_summary_df = pd.DataFrame({"School Type": schoolType,
      "Total Students": totalStudents,
      "Total School Budget": totalSchoolBudget,
      "Per Student Budget": perStudentBudget,
      "Average Math Score": MathScoreaverage,
      "Average Reading Score": ReadingScoreaverage,
      "% Passing Math": PassingMathpercentage,
      "% Passing Reading": PassingReadingpercentage,
      "% Overall Passing": OverallPassingPercentage})

# Sort and display the top prefoming schools
school_summary_df = school_summary_df.sort_values(["% Overall Passing"], ascending=False)

school_summary_df["Total School Budget"] = school_summary_df["Total School Budget"].map("${:,.2f}".format)
school_summary_df["Per Student Budget"] = school_summary_df["Per Student Budget"].map("${:,.2f}".format)

school_summary_df[["School Type",
      "Total Students",
      "Total School Budget",
      "Per Student Budget",
      "Average Math Score",
      "Average Reading Score",
      "% Passing Math",
      "% Passing Reading",
      "% Overall Passing"]].head()
```
